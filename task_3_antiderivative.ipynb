```python
class AntiderivativeTrapeziumAbstract:
       
    def __init__(self,f, xLeft, numBlocks):
        self.f = f
        self.xLeft = xLeft # xLeft = a
        self._numBlocks = numBlocks
        
        self.alpha = alpha
        self.dots = dots
        
        print('Пустой объект класса ' + self.__class__.__name__ + ' создан.')
 
    def __call__(self, x):
            x = float(x)
            gridPoints = np.linspace(self.xLeft, x, self._numBlocks + 1)
            gridLength = len(gridPoints)
            h = (x - self.xLeft)/self._numBlocks
            
            der = 0
            for i, k in zip(self.alpha, self.dots):
                der = der + i*self.f(self.xLeft+k*h)*h
            return der
```

```python
class Antiderivative1(AntiderivativeTrapeziumAbstract):
    def __init__(self, f, xLeft, numBlocks):
        self.f = f
        self.xLeft = xLeft # xLeft = a
        self._numBlocks = numBlocks
        
        self.alpha = np.append([0.],np.ones(self._numBlocks,dtype = float))
        self.dots = np.arange(0.,self._numBlocks+1., 1.)
        
class Antiderivative2(AntiderivativeTrapeziumAbstract):
    def __init__(self, f, xLeft, numBlocks):
        self.f = f
        self.xLeft = xLeft # xLeft = a
        self._numBlocks = numBlocks
        
        self.alpha = np.append(np.ones(self._numBlocks,dtype = float),[0.])
        self.dots = np.arange(0.,self._numBlocks+1., 1.)
        
class Antiderivative3(AntiderivativeTrapeziumAbstract):
    def __init__(self, f, xLeft, numBlocks):
        self.f = f
        self.xLeft = xLeft # xLeft = a
        self._numBlocks = numBlocks
        
        self.alpha = np.append(np.ones(self._numBlocks,dtype = float),[0.])
        self.dots = np.arange(0.,self._numBlocks+1., 1.) + 0.5
        
class Antiderivative4(AntiderivativeTrapeziumAbstract):
    def __init__(self, f, xLeft, numBlocks):
        self.f = f
        self.xLeft = xLeft # xLeft = a
        self._numBlocks = numBlocks
        
        self.alpha = np.append([0.5],np.append(np.ones(self._numBlocks,dtype = float), [0.5]))
        self.dots = np.arange(0.,self._numBlocks+1., 1.) + 0.5
```

```python
x0 = 2.
xSym = smp.Symbol('x')
fSym = smp.sin(xSym**2)
F = Antiderivative1( f=smp.lambdify(xSym, fSym), \
                                  xLeft=0., numBlocks=100)
F(x0)
print ( 'значение в точке %.2f: %.2f ' % ( x0, F(x0)))
F = Antiderivative2( f=smp.lambdify(xSym, fSym), \
                                  xLeft=0., numBlocks=100)
F(x0)
print ( 'значение в точке %.2f: %.2f ' % ( x0, F(x0)))
F = Antiderivative3( f=smp.lambdify(xSym, fSym), \
                                  xLeft=0., numBlocks=100)
F(x0)
print ( 'значение в точке %.2f: %.2f ' % ( x0, F(x0)))
F = Antiderivative4( f=smp.lambdify(xSym, fSym), \
                                  xLeft=0., numBlocks=100)
F(x0)
print ( 'значение в точке %.2f: %.2f ' % ( x0, F(x0)))
```
